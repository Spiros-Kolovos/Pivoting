```python
# Transformation of sheet TOTAL
df2 = pd.read_excel("COUNTRY_+NUTS2_01_01_1990_2023_1y.xlsx", sheet_name='TOTAL', index_col=None, skiprows=5)

for k in range(17):
    df1 = pd.read_excel("COUNTRY AND NUTS2_EUROSTAT.xlsx", sheet_name=f'Sheet {k+1}', index_col=None, skiprows=8, header=1)
    # Remove columns that contain "Unnamed" in their names
    df1 = df1.loc[:, ~df1.columns.str.contains('^Unnamed:')]
    df1 = df1[:-5] # delete 5 last rows    
    
    for j in range(2, df1.shape[1]):
        for i in range(2, len(df1)):
            df2.iloc[107*(j-2)+i+1,k+1] = df1.iloc[i,j]
        #df2.iloc[i+3, 1] = df1.iloc[i+2,2]
        
# Transformation of sheet MALE
df3 = pd.read_excel("COUNTRY_+NUTS2_01_01_1990_2023_1y.xlsx", sheet_name='MALE', index_col=None, skiprows=5)

for k in range(17,34):
    df1 = pd.read_excel("COUNTRY AND NUTS2_EUROSTAT.xlsx", sheet_name=f'Sheet {k+1}', index_col=None, skiprows=8, header=1)
    # Remove columns that contain "Unnamed" in their names
    df1 = df1.loc[:, ~df1.columns.str.contains('^Unnamed:')]
    df1 = df1[:-5] # delete 5 last rows    
    
    for j in range(2, df1.shape[1]):
        for i in range(2, len(df1)):
            df3.iloc[107*(j-2)+i+1,k+1-17] = df1.iloc[i,j]
        #df2.iloc[i+3, 1] = df1.iloc[i+2,2]

# Transformation of sheet FEMALE        
df4 = pd.read_excel("COUNTRY_+NUTS2_01_01_1990_2023_1y.xlsx", sheet_name='FEMALE', index_col=None, skiprows=5)

for k in range(34,51):
    df1 = pd.read_excel("COUNTRY AND NUTS2_EUROSTAT.xlsx", sheet_name=f'Sheet {k+1}', index_col=None, skiprows=8, header=1)
    # Remove columns that contain "Unnamed" in their names
    df1 = df1.loc[:, ~df1.columns.str.contains('^Unnamed:')]
    df1 = df1[:-5] # delete 5 last rows    
    
    for j in range(2, df1.shape[1]):
        for i in range(2, len(df1)):
            df4.iloc[107*(j-2)+i+1,k+1-34] = df1.iloc[i,j]
        #df2.iloc[i+3, 1] = df1.iloc[i+2,2]
        
```

```python
# Transformation of sheet ΣΥΝΟΛΟ
df2 = pd.read_excel("NUTS3_P_2021_2023_5y.xlsx", sheet_name='ΣΥΝΟΛΟ', index_col=None, skiprows=4)

for k in range(56):
    df1 = pd.read_excel("NUTS3_EUROSTAT.xlsx", sheet_name=f'Sheet {k+1}', index_col=None, skiprows=8, header=1)
    # Remove columns that contain "Unnamed" in their names
    df1 = df1.loc[:, ~df1.columns.str.contains('^Unnamed:')]
    df1 = df1[:-5] # delete 5 last rows    
    df1 = df1[~df1.apply(lambda row: row.astype(str).str.contains('Unknown')).any(axis=1)]
    
    for j in range(2, df1.shape[1]):
        for i in range(2, len(df1)):
            df2.iloc[25*(j-2)+i+2,k+1] = df1.iloc[i,j]
        #df2.iloc[i+3, 1] = df1.iloc[i+2,2]
        
# Transformation of sheet ΑΝΔΡΕΣ
df3 = pd.read_excel("NUTS3_P_2021_2023_5y.xlsx", sheet_name='ΑΝΔΡΕΣ', index_col=None, skiprows=4)

for k in range(56,2*56):
    df1 = pd.read_excel("NUTS3_EUROSTAT.xlsx", sheet_name=f'Sheet {k+1}', index_col=None, skiprows=8, header=1)
    # Remove columns that contain "Unnamed" in their names
    df1 = df1.loc[:, ~df1.columns.str.contains('^Unnamed:')]
    df1 = df1[:-5] # delete 5 last rows    
    df1 = df1[~df1.apply(lambda row: row.astype(str).str.contains('Unknown')).any(axis=1)]
    
    for j in range(2, df1.shape[1]):
        for i in range(2, len(df1)):
            df3.iloc[25*(j-2)+i+2,k+1-56] = df1.iloc[i,j]
        #df2.iloc[i+3, 1] = df1.iloc[i+2,2]

# Transformation of sheet ΓΥΝΑΙΚΕΣ        
df4 = pd.read_excel("NUTS3_P_2021_2023_5y.xlsx", sheet_name='ΓΥΝΑΙΚΕΣ', index_col=None, skiprows=4)

for k in range(2*56,3*56):
    df1 = pd.read_excel("NUTS3_EUROSTAT.xlsx", sheet_name=f'Sheet {k+1}', index_col=None, skiprows=8, header=1)
    # Remove columns that contain "Unnamed" in their names
    df1 = df1.loc[:, ~df1.columns.str.contains('^Unnamed:')]
    df1 = df1[:-5] # delete 5 last rows    
    df1 = df1[~df1.apply(lambda row: row.astype(str).str.contains('Unknown')).any(axis=1)]
    
    for j in range(2, df1.shape[1]):
        for i in range(2, len(df1)):
            df4.iloc[25*(j-2)+i+2,k+1-2*56] = df1.iloc[i,j]
        #df2.iloc[i+3, 1] = df1.iloc[i+2,2]
```

```python
with pd.ExcelWriter('NUTS3_P_2021_2023_5y_COMPLETED.xlsx', engine='openpyxl') as writer:
    df2.to_excel(writer, sheet_name='ΣΥΝΟΛΟ', index=False)
    df3.to_excel(writer, sheet_name='ΑΝΔΡΕΣ', index=False)
    df4.to_excel(writer, sheet_name='ΓΥΝΑΙΚΕΣ', index=False)
```
